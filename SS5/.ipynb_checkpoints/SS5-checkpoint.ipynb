```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
import tensorflow as tf
import pathlib
```

```python
# OpenCVライブラリで画像を読み込む
img = cv2.imread('train/image/000.png')
label = cv2.imread('train/label/000.png')
# matplotlibライブラリを使って2つの画像を並べて表示
fig, axes = plt.subplots(1, 2)
axes[0].set_axis_off()
axes[0].imshow(img, cmap='gray')
axes[1].set_axis_off()
axes[1].imshow(label*255, cmap='gray')
plt.show()
```

```python
def create_datasets(img_ls, label_ls):
    assert len(img_ls) == len(label_ls), "画像とラベルの数が違うよ!!"
    img_size = 256 # 縦横256px
    
    train_x = np.zeros((len(img_ls), img_size, img_size, 1))
    train_y = np.zeros((len(label_ls), img_size, img_size, 1))
    
    for i in range(len(img_ls)):
        
        assert img_ls[i].stem == label_ls[i].stem ,'画像とラベルが対応してないかも(拡張子含めない名前が違う)'
        
        img = cv2.imread(str(img_ls[i]), cv2.IMREAD_GRAYSCALE)/255
        lab = cv2.imread(str(label_ls[i]), cv2.IMREAD_GRAYSCALE)
        train_x[i,:,:,0] = img
        train_y[i,:,:,0] = lab
    
    return train_x, train_y
```

```python
train_images = sorted(list(pathlib.Path("train/image").iterdir()))
train_labels = sorted(list(pathlib.Path("train/label").iterdir()))
val_images = sorted(list(pathlib.Path("val/image").iterdir()))
val_labels = sorted(list(pathlib.Path("val/label").iterdir()))

train_x, train_y = create_datasets(train_images, train_labels)
test_x, test_y = create_datasets(val_images, val_labels)
```

```python
from keras.layers import Conv2D, Input,  MaxPooling2D, UpSampling2D, Dropout
from keras.models import Model
from keras.layers.merge import concatenate
from keras.optimizers import RMSprop, Adam
from keras import backend as K
```

```python
inputs=Input((256,256,1))

enc1=Conv2D(32, (3, 3), padding='same', activation='relu')(inputs)
enc2=Conv2D(32, (3, 3), padding='same', activation='relu')(enc1)

enc3=MaxPooling2D(pool_size=(2, 2))(enc2)
enc3=Dropout(0.5)(enc3)
enc4=Conv2D(64, (3, 3), padding='same', activation='relu')(enc3)
enc5=Conv2D(64, (3, 3), padding='same', activation='relu')(enc4)
enc6=Conv2D(64, (3, 3), padding='same', activation='relu')(enc5)

enc7=MaxPooling2D(pool_size=(2, 2))(enc6)
enc7=Dropout(0.5)(enc7)
enc8=Conv2D(128, (3, 3), padding='same', activation='relu')(enc7)
enc9=Conv2D(128, (3, 3), padding='same', activation='relu')(enc8)
enc10=Conv2D(128, (3, 3), padding='same', activation='relu')(enc9)

enc11=MaxPooling2D(pool_size=(2, 2))(enc10)
enc11=Dropout(0.5)(enc11)
enc12=Conv2D(256, (3, 3), padding='same', activation='relu')(enc11)
enc13=Conv2D(256, (3, 3), padding='same', activation='relu')(enc12)
enc14=Conv2D(256, (3, 3), padding='same', activation='relu')(enc13)

enc15=MaxPooling2D(pool_size=(2, 2))(enc14)
enc15=Dropout(0.5)(enc15)
enc16=Conv2D(512, (3, 3), padding='same', activation='relu')(enc15)
enc17=Conv2D(512, (3, 3), padding='same', activation='relu')(enc16)
enc18=Conv2D(512, (3, 3), padding='same', activation='relu')(enc17)

dec1=UpSampling2D(size=(2, 2))(enc18)
dec2=concatenate([dec1, enc14], axis=-1)
dec2=Dropout(0.5)(dec2)
dec3=Conv2D(256, (3, 3), padding='same', activation='relu')(dec2)
dec4=Conv2D(256, (3, 3), padding='same', activation='relu')(dec3)

dec5=UpSampling2D(size=(2, 2))(dec4)
dec6=concatenate([dec5, enc10], axis=-1)
dec6=Dropout(0.5)(dec6)
dec7=Conv2D(128, (3, 3), padding='same', activation='relu')(dec6)
dec8=Conv2D(128, (3, 3), padding='same', activation='relu')(dec7)

dec9=UpSampling2D(size=(2, 2))(dec8)
dec10=concatenate([dec9, enc6], axis=-1)
dec10=Dropout(0.5)(dec10)
dec11=Conv2D(64, (3, 3), padding='same', activation='relu')(dec10)
dec12=Conv2D(64, (3, 3), padding='same', activation='relu')(dec11)

dec13=UpSampling2D(size=(2, 2))(dec12)
dec14=concatenate([dec13, enc2], axis=-1)
dec14=Dropout(0.5)(dec14)
dec15=Conv2D(32, (3, 3), padding='same', activation='relu')(dec14)
dec16=Conv2D(32, (3, 3), padding='same', activation='relu')(dec15)

dec17=Conv2D(1, (3, 3), padding='same', activation='sigmoid')(dec16)

model=Model(input=inputs, output=dec17)
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=["accuracy"], optimizer=Adam())
```

```python
batch_size=8
epoch_num=20

history = model.fit(train_x, train_y, batch_size=batch_size, epochs=epoch_num, validation_split = 0.15)
```

```python
#学習曲線
import matplotlib.pyplot as plt

history_dict = history.history
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]

epochs = range(1,len(loss_values)+1)

plt.plot(epochs, loss_values, "bo", label="Training loss")
plt.plot(epochs, val_loss_values, "r", label="Validation loss")
plt.legend()
plt.show()
```

```python
pre = model.predict(test_x)
```

```python
T = np.sum(np.sum(test_y, axis=1), axis=1)
P = np.sum(np.sum(pre, axis=1), axis=1)
```

```python
IoU = TP/(T+P-TP)
```
